```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)
```

```python
df.describe(include = 'all')
```

```python
#Identify Duplicate Rows
df2 = df.drop(df.columns[0],axis=1)
df2.duplicated().sum()
df2.drop_duplicates()
```

```python
df2.columns.tolist()
df2_list = ['Age', 'Employment', 'MainBranch', 'RemoteWork', 'EdLevel', 'YearsCode', 'YearsCodePro', 'OrgSize', 'Country', 'Currency',
 'CompTotal', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith', 'LanguageAdmired', 'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith',
 'DatabaseAdmired', 'PlatformHaveWorkedWith', 'PlatformWantToWorkWith', 'PlatformAdmired', 'WebframeHaveWorkedWith', 'WebframeWantToWorkWith',
 'WebframeAdmired',  'EmbeddedHaveWorkedWith', 'EmbeddedWantToWorkWith', 'EmbeddedAdmired', 'WorkExp', 'Industry', 'ConvertedCompYearly', 'JobSat']
df3 = df2[df2_list]
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
## Write your code here
# Visualize missing values as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df3.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
df3['Age'].replace("Prefer not to say","25-34 years old", inplace = True)
df_age= df3['Age'].value_counts().rename_axis('unique_values').reset_index(name='counts')
```

```python
plt.figure(figsize=(10,7))
plt.pie(df_age['counts'], labels=df_age['unique_values'], autopct='%1.1f%%')
plt.title('Pie Chart of Respondent Age Categories')

plt.show()
```

```python
cat_dict = {"Bachelor":"Bachelors",
            "Master": "Masters",
            "Some college/university study without earning a degree": "College/university Dropout",
            "Secondary school": "Secondary school",
            "Professional degree": "Professional degree",
            "Associate degree":"Associate degree",
            "Primary/elementary school": "Primary school",
            "Something else":"Something else"
            }
df3['Education'] =  (df3['EdLevel'].str.extract(fr"\b({'|'.join(cat_dict.keys())})\b")[0]
                    .map(cat_dict)
                 )
df3['Education'].replace("Something else",np.nan, inplace = True)
```

```python
df_age_edu=pd.crosstab(df3['Age'],df3['Education']).reset_index()
df_age_edu
```

```python
# plot grouped bar chart
df_age_edu.plot(x='Age',
        kind='bar',
        stacked=False,
        title='Grouped Bar chart for Education with Respondent Age Categories')
```
